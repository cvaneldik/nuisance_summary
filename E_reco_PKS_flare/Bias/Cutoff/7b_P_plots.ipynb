```python
results = dict()

parameters = ['amplitude', 'cutoff', 'lambda_', 'index']
valuies = [amplitude, cutoffs, lambda_, indices]



for i, p in enumerate(parameters):
    subresult = dict()
    subresult['values'] = valuies[i]
    if p == 'cutoff':
        subresult['values_asimov'] = [1/dataset_asimov.models[0].parameters['lambda_'].value, 
                                       dataset_asimov.models[0].parameters['lambda_'].error/ dataset_asimov.models[0].parameters['lambda_'].value**2]
        subresult['values_asimov_N'] = [1/dataset_asimov_N.models[0].parameters['lambda_'].value, 
                                       dataset_asimov_N.models[0].parameters['lambda_'].error/ dataset_asimov_N.models[0].parameters['lambda_'].value**2]
    else:
        subresult['values_asimov'] = [dataset_asimov.models[0].parameters[p].value, 
                                  dataset_asimov.models[0].parameters[p].error]
        subresult['values_asimov_N'] = [dataset_asimov_N.models[0].parameters[p].value, 
                                  dataset_asimov_N.models[0].parameters[p].error]
        
    within = []
    within_N= []
    for n in subresult['values']:
        larger = np.all(n[0] >= (subresult['values_asimov'][0] - subresult['values_asimov'][1]))
        smaller = np.all(n[0] <= (subresult['values_asimov'][0] + subresult['values_asimov'][1]))
        within.append(larger and smaller)

        larger = np.all(n[0] >= (subresult['values_asimov_N'][0] - subresult['values_asimov_N'][1]))
        smaller = np.all(n[0] <= (subresult['values_asimov_N'][0] + subresult['values_asimov_N'][1]))
        within_N.append(larger and smaller)
        
        
    within_total = [np.count_nonzero(within)/N,
              np.count_nonzero(within_N)/N]
    subresult['within'] = within_total
    
    results[p] = subresult
        
```
